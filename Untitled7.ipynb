```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10,6)

# Carregar dados
url = "https://raw.githubusercontent.com/ingridcristh/challenge2-data-science/refs/heads/main/TelecomX_Data.json"
df = pd.read_json(url)

# Converter colunas complexas (dict ou list) para string
for col in df.columns:
    if df[col].apply(lambda x: isinstance(x, (dict, list))).any():
        df[col] = df[col].astype(str)

# Forçar colunas numéricas
numeric_cols_to_convert = ['tenure', 'Charges.Monthly', 'Charges.Total']
for col in numeric_cols_to_convert:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')  # valores inválidos viram NaN

# Detectar coluna de churn
churn_col = [col for col in df.columns if 'churn' in col.lower()]
churn_col = churn_col[0] if churn_col else None

# Separar colunas categóricas e numéricas
categorical_cols = df.select_dtypes(include=['object', 'bool']).columns.tolist()
if churn_col in categorical_cols:
    categorical_cols.remove(churn_col)
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()

# Plot churn
if churn_col:
    sns.countplot(x=churn_col, data=df)
    plt.title(f"Distribuição de {churn_col}")
    plt.show()

# Análise categórica
for col in categorical_cols:
    plt.figure(figsize=(8,4))
    sns.countplot(x=col, hue=churn_col, data=df)
    plt.title(f"{col} vs {churn_col}")
    plt.xticks(rotation=45)
    plt.show()

# Análise numérica
for col in numerical_cols:
    plt.figure(figsize=(8,4))
    sns.boxplot(x=churn_col, y=col, data=df)
    plt.title(f"{col} vs {churn_col}")
    plt.show()

# Correlação entre numéricas
if numerical_cols:
    plt.figure(figsize=(10,8))
    sns.heatmap(df[numerical_cols].corr(), annot=True, fmt=".2f", cmap="coolwarm")
    plt.title("Correlação entre variáveis numéricas")
    plt.show()
else:
    print("Não há colunas numéricas para correlação.")

# Salvar dataset tratado
df.to_csv("TelecomX_Tratado.csv", index=False)
print("Dataset tratado salvo como 'TelecomX_Tratado.csv'.")
```
